```python
from sqlalchemy import create_engine, Column, Integer, String, Numeric, Text, Float
from sqlalchemy.orm import Session
from sqlalchemy.ext.automap import automap_base

import pandas as pd
```

```python
# Create the engine to connect SQLAlchemy to SQLite
engine = create_engine("sqlite:///demographics.sqlite")

# Reflect database into ORM class
Base = automap_base()
Base.prepare(engine, reflect=True)
Demographics = Base.classes.demographics

# Start a session to query the database
session = Session(engine)
```

```python
demo_df = pd.read_sql(session.query(Demographics).statement, engine, index_col="id")
demo_df
```

```python
demo_df.loc[0,"age"] = 60
demo_df
```

```python
demo_df.to_sql("demographics", engine, if_exists="replace")
```
