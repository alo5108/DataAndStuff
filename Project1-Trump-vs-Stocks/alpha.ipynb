```python
import json
import pandas as pd
import numpy as np
import requests as req
import matplotlib.pyplot as plt
```

```python

url_iex='https://api.iextrading.com/1.0/ref-data/symbols'
```

```python
response=req.get(url_iex).json()
```

```python
print(len(response))
```

```python
comp_name=[]
comp_symbol=[]


for x in response:
        name=x['name']
        symbol=x['symbol']
        
        
        comp_name.append(name)
        comp_symbol.append(symbol)
        
```

```python
Companies_info=pd.DataFrame({'Company Name':comp_name,'Company Stock Symbol':comp_symbol})
Companies_info
```
